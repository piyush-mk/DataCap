```python
import os
import h5py
from datetime import datetime, timedelta
import matplotlib.pyplot as plt

# assign directory
directory = './ndvi_data'
i = 0
for filename in (os.listdir(directory)):
    f = os.path.join(directory, filename)
    julian_date = str(f)[-16:-9]
    date = datetime.strptime(julian_date, '%Y%j')
    if os.path.isfile(f):
        with h5py.File(f, 'r') as file:
            data = file['HDFEOS']['GRIDS']['VIIRS_Grid_16Day_VI_1km']['Data Fields']['1 km 16 days NDVI'][:]

        plt.figure() 
        plt.imshow(data)
        plt.colorbar(label='NDVI')
        plt.title(f"Date of acquisition: {date.strftime('%Y-%m-%d')}")
        plt.grid(False)
        plt.axis('off')
        n = 'ndvi_image/ndvi'+str(i)+'.jpg'
        i+=1
        plt.show()
```

```python
import os
import h5py
from datetime import datetime
import matplotlib.pyplot as plt

# assign directory
directory = './ndvi_data'

# iterate over files in that directory
for filename in sorted(os.listdir(directory)):
    f = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(f):
        with h5py.File(f, 'r') as file:
            # Navigate through groups to the desired dataset
            data = file['HDFEOS']['GRIDS']['VIIRS_Grid_16Day_VI_1km']['Data Fields']['1 km 16 days NDVI'][:]

        # Extract the timestamp from the filename
        timestamp = filename.split('.')[5]
    # Extract the year and day of year from the filename
        year = int(filename[10:14])
        day_of_year = float(filename[14:16])  # Extract only the numeric part

        # Compute the date
        date = datetime(year, 1, 1) + timedelta(int(day_of_year) - 1)

        # Visualize the data
        plt.figure()  # added this line to create a new figure for each image
        plt.imshow(data)
        plt.colorbar(label='NDVI')
        plt.title(f"Date of acquisition: {date}")
        plt.grid(False)
        plt.axis('off')
        plt.show()
```
