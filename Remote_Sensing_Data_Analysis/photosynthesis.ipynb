```python
from netCDF4 import Dataset
import netCDF4
import matplotlib.pyplot as plt
from datetime import datetime
import os
import xarray as xr
```

```python
directory = './photosynthesis_data'
j = 0

for i, filename in enumerate(os.listdir(directory)):
    f = os.path.join(directory, filename)
    ds = xr.open_dataset(f, engine='netcdf4')
    julian_date = str(f)[-17:-10]
    date = datetime.strptime(julian_date, '%Y%j')
    print(date.strftime('%Y-%m-%d'))
    
    data = ds['GMT_0600_PAR']
    plt.figure(figsize=(6, 4))
    
    # Plot the data    6 a.m.
    plt.imshow(data, cmap='jet')
    plt.xlabel('XDim:MODISRAD')
    plt.ylabel('YDim:MODISRAD')
    plt.title('Total PAR at 06:00')

    # Show the plot
    n = 'photosynthesis_image/pic'+str(i)+'.jpg'
    i+=1
    plt.savefig(n)
    
    data = ds['GMT_0900_PAR']

    # Create a new figure for each image
    plt.figure(figsize=(6, 4))

    
    # Plot the data    9 a.m.
    plt.imshow(data, cmap='jet')
    plt.xlabel('XDim:MODISRAD')
    plt.ylabel('YDim:MODISRAD')
    plt.title('Total PAR at 09:00')

    # Show the plot
    n = 'photosynthesis_image/pic'+str(i)+'.jpg'
    i+=1
    plt.savefig(n)
    
    data = ds['GMT_1200_PAR']

    # Create a new figure for each image
    plt.figure(figsize=(6, 4))

    # Plot the data    12 p.m.
    plt.imshow(data, cmap='jet')
    plt.xlabel('XDim:MODISRAD')
    plt.ylabel('YDim:MODISRAD')
    plt.title('Total PAR at 12:00')

    # Show the plot
    n = 'photosynthesis_image/pic'+str(i)+'.jpg'
    i+=1
    plt.savefig(n)
    
    data = ds['GMT_1500_PAR']

    # Create a new figure for each image
    plt.figure(figsize=(6, 4))

    # Plot the data    3 p.m.
    plt.imshow(data, cmap='jet')
    plt.xlabel('XDim:MODISRAD')
    plt.ylabel('YDim:MODISRAD')
    plt.title('Total PAR at 15:00')

    # Show the plot
    n = 'photosynthesis_image/pic'+str(i)+'.jpg'
    i+=1
    plt.savefig(n)
    
    data = ds['GMT_1800_PAR']

    # Create a new figure for each image
    plt.figure(figsize=(6, 4))

    # Plot the data    6 p.m.
    plt.imshow(data, cmap='jet')
    plt.xlabel('XDim:MODISRAD')
    plt.ylabel('YDim:MODISRAD')
    plt.title('Total PAR at 18:00')

    # Show the plot
    n = 'photosynthesis_image/pic'+str(i)+'.jpg'
    i+=1
    plt.show()
    break
    
```
